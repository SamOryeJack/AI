```python
import fitz  # PyMuPDF
import spacy
from spacy import displacy
from google.colab import drive


# Mount Google Drive and load the dataset
drive.mount('/content/drive')
# Specify file path
mypdf = '/content/drive/MyDrive/Job/Github/Text_Mining/tax-efficient-withdrawal-strategies.pdf'
# Open the PDF file
pdf_document = fitz.open(mypdf)

# Get the number of pages in the PDF file
page_count = len(pdf_document)

print(page_count)
```

```python
# Step 1: Extract text from each page
output = []
for page_num in range(page_count):
    page = pdf_document.load_page(page_num)
    output.append(page.get_text())

# Concatenate the extracted text
all_texts = ' '.join(output)

# Close the PDF document
pdf_document.close()

# Load the spaCy English model (you might need to download it first)
nlp = spacy.load("en_core_web_sm")
```

```python
# Step 2: Extract all the tokens and lemmas from the texts
doc = nlp(all_texts)
```

```python
# Step 3: Remove default stop words in SpaCy
filtered_tokens = [token.text for token in doc if not token.is_stop]
```

```python
# Step 4: Customize the stop words in SpaCy
custom_stop_words = {"tax", "account"}

# Add custom stop words to spaCy's default stop word list
for token in custom_stop_words:
    nlp.Defaults.stop_words.add(token)

# Set the 'is_stop' indicator for custom stop words in spaCy's vocabulary
for token in custom_stop_words:
    nlp.vocab[token].is_stop = True

# Remove customized default stop words from the filtered tokens
filtered_tokens = [token for token in filtered_tokens if token not in custom_stop_words]
```

```python
# Step 5: Perform part-of-speech tagging
pos_tags = [(token.text, token.pos_) for token in doc]
```

```python
# Step 6: Visualize the dependency parser
# Iterate through tokens and print their text and POS tags
for token in doc:
    print(f"Token: {token.text}, POS Tag: {token.pos_}")
```

```python
# Step 7: Perform named entity recognition (NER)
ner_results = [(ent.text, ent.label_) for ent in doc.ents]
```

```python
# set person to mark and color to use
option2 = {"ents": ['MONEY'], "colors": {'MONEY':'#82E0AA'}}
option3 = {"ents": ['QUANTITY'], "colors": {'QUANTITY':'#e082b8'}}
## Visualize it by seeting style to be "ent" and jupyter to be True and the corresponding options
displacy.render(doc, style="ent", jupyter = True, options=option2)
displacy.render(doc, style="ent", jupyter = True, options=option3)
```

```python
# Step 8: Visualize MONEY and QUANTITY entities
money_quantity_entities = {"ents": ['MONEY', 'QUANTITY'], "colors": {'MONEY':'#55FF55', 'QUANTITY':'#5555FF'}}
displacy.render(doc, style="ent", jupyter=True, options=money_quantity_entities)
```
